### Exercício 4)

O DataFrame `enem` possui valores faltantes na coluna 'Classificacao'. Preencha os valores faltantes com o valor da 'moda' da coluna. Qual o valor da moda utilizado para preencher os valores faltantes?

a) 1<br>
b) <font color='red'>2</font><br>
c) 3<br>
d) 4

```python
mask = enem['Classificacao'].isna()
print('Numero de valores faltantes: ',mask.sum())

moda = enem['Classificacao'].mode()
enem['Classificacao'].fillna(moda[0])

print('Valor da moda: ',moda[0])
```
